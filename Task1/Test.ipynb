```python

# (Opcional) si no est√° en el PATH:
pytesseract.pytesseract.tesseract_cmd = r"C:/Program Files/ts/tesseract.exe"
os.environ["TESSDATA_PREFIX"] = r"C:\Program Files\ts\tessdata"
```

```python

# Ruta local de la imagen
image_path = "captura1.png"  # reemplaza por el nombre de tu imagen

# Cargar imagen
image = Image.open(image_path)

```

```python
# Mostrar imagen (opcional)
plt.imshow(image)
plt.axis('off')
plt.show()
```
